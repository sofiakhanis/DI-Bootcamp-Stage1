```python
import matplotlib.pyplot as plt
import pandas as pd

x = [1, 2, 3, 4]
y = [10, 20, 0, 40]

plt.plot(x, y, color = 'red', linestyle = "dotted", linewidth = 2, marker = 'o', markersize = 6)
plt.show()
```

```python
plt.bar (['kevin', 'rodrigo', 'eva', 'vince'], [99, 44, 2, 20], color = ["green", "yellow", "blue", "red"], edgecolor = "black", hatch = "/")
plt.title ("DnD's risky situations")
plt.xlabel ("characters")
plt.ylabel ("situations")
plt.show()
```

```python
import numpy as np

data = np.random.normal (0, 1, 1000)
plt.hist(data, bins = 10, color = "black", edgecolor = "red")
plt.title ("Changes")
plt.ylabel("Minus")
plt.xlabel ("Plus")
plt.show()
```

```python
explode = (0.1, 0.1, 0.1)
plt.pie ([10, 30, 40], labels = ["Dior", "Chanel", "Burberry"], colors = ["blue", "gold", "beige"], explode = explode)
plt.title ("Brands")
plt.show()
```

```python
data1 = np.random.randn (100)
plt.boxplot(data1)
plt.title ("Basic box plot")
plt.ylabel ("Values")
plt.show()
```

```python
import seaborn as sns

data = sns.load_dataset ('tips')
sns.histplot (data['total_bill'], bins = 100, color = "red")
plt.title ('Histogram')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], shade = True, color = 'purple')
plt.title ('KDE of Total Bill')
plt.show()
```
